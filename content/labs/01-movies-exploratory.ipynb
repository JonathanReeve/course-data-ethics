# Lab 1: An Exploratory Data Analysis using the Movies Dataset

Let's start by loading the Movies dataset from vega_datasets. 

```python
movies = vega_datasets.data.movies()
```

What columns do we have access to? 

```python
movies.columns
```

What's the distribution of genres? 

```python
movies['Major_Genre'].hist(xrot=45)
```

## Is there a correlation between MPAA rating and the Rotten Tomatoes rating? 

```python
movies.groupby('MPAA_Rating')['Rotten_Tomatoes_Rating'].mean().plot(kind='bar')
```

What about the source (novel / short story) of the script? 

```python
movies.groupby('Source')['Rotten_Tomatoes_Rating'].mean().plot(kind='bar')
```

```python
movies[movies['Source'] == 'Based on Comic/Graphic Novel']['MPAA_Rating'].hist()
```

```python
movies[movies['Source'] == 'Based on Book/Short Story']['MPAA_Rating'].hist()
```
